```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
start_date = '2023-04-01'
end_date = '2024-03-31'
data = yf.download( 'TCS.NS', start = start_date, end = end_date)
```

```python
data['SMA 10'] = np.nan
for i in range(9,data.shape[0]):
    data['SMA 10'].iloc[i] = (data['Close'].iloc[i-9:i+1]).mean()
data['SMA 20'] = data['Close'].rolling(20).mean()
data
```

```python
plt.plot(data['Close'] , label = 'Closing Price')
plt.plot(data['SMA 10'] , label = 'SMA 10')
plt.plot(data['SMA 20'] , label = 'SMA 20')
plt.xlabel('Time')
plt.legend()
plt.show()
```
